# 30. Logistic Regression (Practical) (Binary Classification) (Polynomial Input))

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r'Data/Social_Network_Ads_2.csv')
dataset.head(3)
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="Age", y="EstimatedSalary", data=dataset, hue="Purchased")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
x
```

```python
y=dataset["Purchased"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=lr)
plt.show()
```

#### You can see the data is not linearly separable. so we will classify this data through **polynomial Feature**

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=3)
pf.fit(x)
pf.transform(x)
```

The output data is in array form so we will convert it into dataset format

```python
x = pd.DataFrame(pf.transform(x))
x.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lrg = LogisticRegression()
lrg.fit(x_train, y_train)
```

```python
lrg.score(x_test, y_test)*100
```
