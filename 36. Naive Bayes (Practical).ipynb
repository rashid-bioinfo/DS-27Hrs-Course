# 36. Naive Bayes (Practical)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r'Data/placement_3.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

**To check if the data is normally distributed or not, we will use disribution plot to check this**

```python
sns.kdeplot(data=dataset["cgpa"])
plt.show()
```

```python
sns.kdeplot(data=dataset["cgpa"])
plt.show()
```

#### So will apply Gaussian Naive Bayes, b/c data is normally distributed.

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa", y="score", data=dataset, hue="placed")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
x
```

```python
y = dataset["placed"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
gnb.score(x_test, y_test)*100, gnb.score(x_train, y_train)*100
```

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
gnb.score(x_test, y_test)*100, gnb.score(x_train, y_train)*100
```

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
gnb.score(x_test, y_test)*100, gnb.score(x_train, y_train)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=gnb)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=mnb)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=bnb)
plt.show()
```
