## 12.1 Label encoding on Nominal data

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.DataFrame({'name':['Rashid', 'Lion', 'Computer', 'Gym', 'Plant']})
df
```

```python
le = LabelEncoder()
df['en_name'] = le.fit_transform(df['name'])
```

## 12.1 Label encoding on Ordinal data

### 12.1.1 Ordincal encoding through Cyclic Line

```python
dfo = pd.DataFrame({'size': ['s', 'm', 'l', 'xl', 'xxl', 's', 's', 's', 'xl', 'm', 'l']})
dfo.head(3)
```

```python
ord_data = [['s', 'm', 'l', 'xl', 'xxl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# oe = OrdinalEncoder() : This will encode the data alphabatically
oe = OrdinalEncoder(categories=ord_data)
oe.fit(dfo[['size']])
```

```python
oe.transform(dfo[['size']])
```

```python
dfo['size_en'] = oe.transform(dfo[['size']])
dfo
```

### 12.1.2 Ordincal encoding through Map function

```python
# In map function, you can manually assign numbers to each data type, for example
# You can assign any number
ord_data1 = {'s':0, 'm':1, 'l':2, 'xl':3, 'xxl':4}
```

```python
dfo['size'].map(ord_data1)
```

```python
dfo['size_en_map'] = dfo['size'].map(ord_data1)
dfo
```

### 12.2 Perform Ordinal Encoding on big data

```python
dataset = pd.read_csv('Data\loan.csv')
```

```python
dataset.head(3)
```

```python
dataset['Property_Area'].unique()
```

```python
# if there is nan (missing data)m, you can fill it
# if the data is categorical then you should do mode filling
dataset['Property_Area'].fillna(dataset['Property_Area'].mode()[0], inplace=True)
```

```python
en_data_loan = [['Urban', 'Rural', 'Semiurban']]
```

```python
oen = OrdinalEncoder(categories=en_data_loan)
oen.fit_transform(dataset[['Property_Area']])
```

```python
dataset['Property_Area'] = oen.fit_transform(dataset[['Property_Area']])
```
