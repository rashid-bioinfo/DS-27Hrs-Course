### 10.2 How to handle missing values (Dropping)

Deleting in 2 ways:
1) If a column contains 50% missing value, then delete the whole column
2) Only delete the rows which are having missing values, instead of deleting whole column

#### Deleting a column from the data

```python
dataset.isnull().sum()
```

We will delete Credit_History column as it contains more missing values

```python
# Inplace function will enable to make changes in the original datasheet that is dataset.
# It will write changes in the excisting file i.e., load.csv
dataset.drop(columns=['Credit_History'], inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

#### Deleting rows containing null values

```python
# Again we use inplace function to write the changes in the same datasheet instead of creating new data sheet
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
sns.heatmap(dataset.isnull())
plt.show()
```

```python
dataset.shape
```

To check how much data has been dropped (deleted)

```python
((614-523)/614)*100
```

14% data is lost

## 10.3 Handling Missing Values (Imputing Category Data)

While dropping the data can be harmful as it may contain essential data, so instead of deleting we will fill the data where the missing values are present

We will import the orginial data of loan.csv, as we have dropped missing data and overwrite the changes in the above data

```python
dataset = pd.read_csv('loan.csv')
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
# It will fill all the data using the number provided i.e., 10
# this method is not recommended as it will fill dummy data, lead to wrong insight of data
dataset.fillna(10)
```

So we will fill data wisely, so we will first determine the datatype
String data type is called object data in ML
Data is of two types:
1) Numerical data
2) Categorical data - string data (object type data)
Filling in categorical data:
1) Backward filling
2) Forward filling
3) Mod filling




```python
dataset.info()
```

```python
# Backward filling - Back data will filled , forexample Loan Amount first row is filled with 128, its value is taken from next row data
# nichay wala data oper a k fill ho jaye ga
dataset.fillna(method='bfill')
```

```python
# Forward filling - oper wala data nicha a k fill ho jaye ga 
# by default filling is row wise
dataset.fillna(method='ffill')
```

```python
# by default filling is row wise - So fill data column wise, we will use axis
dataset.fillna(method='ffill', axis=1)
```

- In mode data filling, you will most repeatitive data in missing contents
- fill the missing data in Gender column

#### Fill particular column containing missing value by mode method

```python
dataset['Gender'].mode()
```

```python
dataset['Gender'].mode()[0]
```

```python
dataset['Gender'].fillna(dataset['Gender'].mode()[0], inplace=True)
```

```python
dataset.isnull().sum()
```

#### Fill all columns containing missing value by mode method

1. First you will collect all object datatype

```python
dataset.select_dtypes(include='object')
```

```python
dataset.select_dtypes(include='object').isnull()
```

```python
dataset.select_dtypes(include='object').isnull().sum()
```

```python
dataset.select_dtypes(include='object').columns
```

```python
for i in dataset.select_dtypes(include='object').columns:
    print(i)
```

```python
for i in dataset.select_dtypes(include='object').columns:
    #dataset['Gender'].fillna(dataset['Gender'].mode()[0], inplace=True)
    dataset[i].fillna(dataset[i].mode()[0], inplace=True)
```

```python
dataset.isnull().sum()
```

So all object data has been filled and only numerical data is left to be filled

Import fresh datasheet

#### Find the missing values using scikit learn

```python
from sklearn.impute import SimpleImputer 
```

- sklearn provide variety of options to fill the data. for example fill the data by mean, most fequency (mode), median
- We are now filling the data by using mean method

```python
si = SimpleImputer(strategy='mean')
si.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
```

```python
# To convert this data into csv sheet
si = SimpleImputer(strategy='mean')
arr = si.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
```

```python
new_dataset = pd.DataFrame(arr, columns=dataset.select_dtypes(include='float64').columns)
```

```python
new_dataset.isnull().sum()
```
